```python
'''
StandardScaler는 각 특성의 평균을 0, 분산을 1로 변경하여 모든 특성이 같은 크기를 갖게 합니다. 단, 최소와 최대값의 크기를 제한하지는
않습니다.
RobustScaler는 StandardScaler와 비슷하지만 평균과 분산 대신 중간값과 사분위값으 사용합니다. 따라사 이상치의 영향을 받지 않습니다.
MinMaxScaler는 모든 특성이 0, 1사이에 위치하도록 데이터를 변경합니다. 
Normalizer는 특성 벡터의 유클리디안 길이가 1이 되도록 데이터 포인트를 조정합니다. 데이터의 방향만이 중요할 때 사용합니다.
'''
```

```python
##MinMaxScaler를 이용해서 데이터 전처리를 해보도록 하겠습니다.
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state = 1)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
print("조정 전 최소: \n", X_train.min(axis= 0))
print("조정 전 최대: \n",X_train.max(axis= 0))
print("조정 후 최소: \n", X_train_scaled.min(axis = 0))
print("조정 후 최대: \n",X_train_scaled.max(axis = 0))
```

```python
##MinMaxScaler를 통해 데이터들이 0~1사이 값으로 변한 것을 확인할 수 있습니다.
##테스트 세트도 변환해야합니다.

X_test_scaled = scaler.transform(X_test)
print("조정 후 최소: \n", X_test_scaled.min(axis = 0))
print("조정 후 최대: \n",X_test_scaled.max(axis = 0))
```

```python
###자세히 보시면 X_test를 조정하면 최소가 0, 최대가 1이 아닙니다.
###이는 변화의 비율을 X_train에 맞췄기 때문입니다. 
##데이터를 변활할 때는 훈련 세트에 맞추고 같은 비율로 테스트 세트에 적용해야합니다.
```

```python
##전처리의 유무에 따른 학습률을 확인해보도록 하겠습니다.
from sklearn.svm import SVC

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state = 0)

svm = SVC(C = 100)
svm.fit(X_train, y_train)
print("{:.4f}".format(svm.score(X_test, y_test)))
print('\n')

scaler = MinMaxScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
svm.fit(X_train_scaled, y_train)
print("{:.4f}".format(svm.score(X_test_scaled, y_test)))

##결과가 향상된 것을 확인할 수 있습니다.
```
