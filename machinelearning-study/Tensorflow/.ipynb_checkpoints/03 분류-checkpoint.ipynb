```python
##이향 분류는 정답의 범주가 두 개인 분류 문제입니다. 
##레드 와인과 화이트 와인을 분류하는 문제를 풀어보도록 하겠습니다.
```

```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
##외부에서 데이터 불러오기

red = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv', sep=';')
white = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv', sep=';')
```

```python
##두개의 데이터를 합치기 위해 각 와인임을 알려주는 열을 만들고 두 데이터 프레임을 합쳐야 합니다.
red['type'] = 0
white['type'] =1

wine = pd.concat([red, white])
wine.info()
```

```python
wine.describe()
```

```python
plt.hist(wine['type'])
plt.xticks([0,1])
plt.show()
```

```python
##와인의 수를 보시면 화이트 와인이 훨씬 많은 것을 확인할 수 있습니다.
##이를 훈련셋과 테스트셋으로 나누기 전에 랜덤으로 데이터를 섞을 필요가 있습니다. 
##데이터를 정규화한 후 데이터를 랜덤으로 섞어보도록 하겠습니다.

wine_norm = (wine - wine.min())/(wine.max()-wine.min())
wine_norm.head(3)
```

```python
wine_norm.describe()
```

```python
##판다스 데이터프레임의 sample함수는 frac라는 인자를 받아서 frcc의 비율만큼 행을 랜덤으로 뽑습니다.
wine_shuffle = wine_norm.sample(frac=1)
wine_shuffle.head(10)
```

```python
##to_numpy()로 데이터 프레임을 넘파이 배열로 변환해줍니다.
wine_np = wine_shuffle.to_numpy()
wine_np[:5]
```

```python
##전처리 되고 랜덤으로 섞인 데이터를 훈련셋과 테스트셋으로 나누겠습니다. 
##전체의 0.8만큼을 훈련셋으로 설정하고, Y 값들은 원핫 인코딩으로 바꾸기 위해 tf.keras.utils에서 to_categorical를 사용하겠습니다.

train_idx = int(len(wine_np)*0.8)
X_train, Y_train = wine_np[:train_idx, :-1], wine_np[:train_idx, -1]
X_test, Y_test = wine_np[train_idx: , :-1], wine_np[train_idx: , -1]
Y_train = tf.keras.utils.to_categorical(Y_train, num_classes = 2)
Y_test = tf.keras.utils.to_categorical(Y_test, num_classes = 2)
print(Y_train[0])
print(Y_test[0])
```

```python
##시퀀셜 모델을 통해 딥러닝 학습을 해보도록 하겠습니다.

model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=48, activation='relu', input_shape=(12,)),
    tf.keras.layers.Dense(units=24, activation='relu'),
    tf.keras.layers.Dense(units=12, activation='relu'),
    tf.keras.layers.Dense(units=2, activation='softmax')
])

model.compile(optimizer = tf.keras.optimizers.Adam(lr=0.07), loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
##위 코드를 보시면 회귀에서와 몇가지 다른 점이 있습니다.
##마지막 층의 activation이 softmax인데 이는 결과값으로 확률이 필요한 분야에서 자주 쓰입니다.
##또한 마지막 층의 units가 2인데 이는 정답의 차원수와 동일하게 맞춰줘야 합니다. 이 경우, 정답이 (2,)이므로  units는 2가 됩니다.
##분류문제는 정확도로 퍼포먼스를 측정하기 때문에 accuracy값을 주었습니다.

history = model.fit(X_train,Y_train, epochs=25, batch_size=32, validation_split=0.25)
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.plot(history.history['loss'], 'b-', label='loss')
plt.plot(history.history['val_loss'], 'r--', label='val_loss')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], 'g-', label="accuracy")
plt.plot(history.history['val_accuracy'], 'k--', label='val_accuracy')
plt.ylim(0.7,1)
plt.legend()
plt.show()
```

```python
##다항분류는 범주의 수가 2개를 넘는 겨우입니다.
##앞선 데이터에서 와인의 품질을 다항분류를 통해 예측해보겠습니다.

wine['quality'].value_counts()
```

```python
plt.hist(wine['quality'], bins=7)
```

```python
##데이터의 분포를 보시면 각 데이터의 숫자가 차이가 심하기 때문에 모든 범주에 대한 세세한 분류는 어려울 것 같습니다.
##따라서 3~5는 나쁨, 6은 보통, 7~9는 좋음으로 분류하도록 하겠습니다.
wine.loc[wine['quality']<=5, 'new_quality'] = 0
wine.loc[wine['quality']==6, 'new_quality'] = 1
wine.loc[wine['quality']>=7, 'new_quality'] = 2
wine["new_quality"].value_counts()
```

```python
##다시 데이터를 전처리 하도록 하겠습니다.


wine_norm  = (wine-wine.min())/(wine.max()-wine.min())
wine_shuffle = wine_norm.sample(frac=1)
wine_np = wine_shuffle.to_numpy()

train_idx = int(len(wine_np)*0.8)
X_train, y_train = wine_np[:train_idx, :-1], wine_np[: train_idx, -1]
X_test, y_test = wine_np[train_idx:, :-1], wine_np[train_idx:, -1]
y_train = tf.keras.utils.to_categorical(y_train, num_classes=3)
y_test = tf.keras.utils.to_categorical(y_test, num_classes=3)
```

```python
wine.info()
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=48, activation="relu", input_shape=(12,)),
    tf.keras.layers.Dense(units=24, activation="relu"),
    tf.keras.layers.Dense(units=12, activation="relu"),
    tf.keras.layers.Dense(units=3, activation="softmax")
])
model.compile(optimizer=tf.keras.optimizers.Adam(lr=0.07), loss="categorical_crossentropy", metrics=["accuracy"])

history=model.fit(X_train, y_train, epochs=25, batch_size=32, validation_split=0.25)

##이진 분류일때보다는 성능이 떨어지지만 80%를 넘는 준수한 예측 성적을 보여주었습니다.
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.plot(history.history['loss'], 'b-', label='loss')
plt.plot(history.history['val_loss'], 'r--', label='val_loss')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], 'g-', label="accuracy")
plt.plot(history.history['val_accuracy'], 'k--', label='val_accuracy')
plt.ylim(0.7,1)
plt.legend()
plt.show()
```

```python
##Fashion MNIST데이터를 이용하여 딥러닝 모델을 만들어보도록 하겠습니다.
fashion_mnist = tf.keras.datasets.fashion_mnist
(X_train, y_train),(X_test, y_test) = fashion_mnist.load_data()
```

```python
X_train.shape ## 데이터셋은 28x28의 픽셀을 가지고 있는 이미지를 총 60000개 가지고 있습니다. 3차원 데이터셋입니다.
```

```python
##정규화를 진행하도록 하겠습니다.
##여기서 주의할 점은 Y에 대해서는 원핫 인코딩을 하지 않는데 이 경우는 0이 너무 많은 희소 행렬이기떄문입니다.
##이런 희소행렬의 경우는 원핫 인코딩 보다는 숫자로 표현하는게 더 효율적이고 이미 그런식으로 데이터게 설정돼 있기 때문에
#따로 변활할 필요가 없습니다.

X_train =X_train/255.0
X_test = X_test/ 255.0
```

```python
##Sequential모델을 만들 때 변화된 점은 원핫 인코딩이 아닌 정답 행렬을 처리하기 위해 compile에서 loss의 인수를 바꿔야합니다.
##이 경우는 희소행렬을 의마하는 sparse를 붙여서 sparse_categorical_crossentropy를 써줘야합니다.
##또한 원본이 2차원 array이기 때문에 Flatten레이어를 사용하는 이는 다차원 데이터를 1차원으로 정렬하는 역할을 합니다.
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(units=128, activation="relu"),
    tf.keras.layers.Dense(units=10, activation="softmax")
])
model.compile(optimizer = tf.keras.optimizers.Adam(), loss="sparse_categorical_crossentropy",  metrics=["accuracy"])

model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=25, validation_split=0.25)
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history['loss'], 'b-', label="loss")
plt.plot(history.history['val_loss'], 'r--', label="val_loss")
plt.legend()


plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], 'g-', label="accuracy")
plt.plot(history.history["val_accuracy"], 'k--', label="val_accuracy")
plt.legend()
plt.show()
```
