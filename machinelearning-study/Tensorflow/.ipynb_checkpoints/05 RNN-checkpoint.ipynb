```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
##SimpleRNN
X = []
Y = []

for i in range(6):
    lst = list(range(i, i+4))
    X.append(list(map(lambda c: [c/10], lst)))
    Y.append((i+4)/10)
    
X = np.array(X)
Y = np.array(Y)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.SimpleRNN(units=10, return_sequences = False, input_shape=[4,1]),
    tf.keras.layers.Dense(1)
])
model.compile(optimizer='adam', loss="mse")
model.summary()


##여기서 input_shape = [4,1]에서 4는 timesteps, 1은 input_dim을 나타냅니다.
##timesteps는 순환 신경망이 입력에 대한 계산을 반복하는 횟수이고 input_dim은 입력 벡터의 크기를 나타냅니다.
```

```python
model.fit(X,Y, epochs=100, verbose=0)
```

```python
print(model.predict(X))
```

```python
##LSTM
##비교하기위해 우선 SimpleRNN으로 먼저 학습하겠습니다.
X = []
Y = []
for i in range(3000):
    lst = np.random.rand(100)
    idx = np.random.choice(100,2,replace = False)
    zeros = np.zeros(100)
    zeros[idx] = 1
    X.append(np.array(list(zip(zeros, lst))))
    Y.append(np.prod(lst[idx]))
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.SimpleRNN(units=30, return_sequences = True, input_shape = [100,2]),
    tf.keras.layers.SimpleRNN(units=30),
    tf.keras.layers.Dense(1)
])
model.compile(optimizer='adam', loss='mse')
model.summary()

##return_sequences=True는 레이어의 출력을 다음 레이어로 그대로 넘겨주라는 뜻입니다.
```

```python
X=np.array(X)
Y = np.array(Y)
```

```python
history=model.fit(X[:2560], Y[:2560], epochs=100, validation_split=0.25)
```

```python
##훈련 데이터 손실인 loss는 점차 감소하지만 val_loss는 증가하여 과적합이 발생했음을 알 수 있습니다.
##이를 시각화하면 다음과 같습니다.
plt.plot(history.history["loss"], 'bo', label="loss")
plt.plot(history.history["val_loss"], 'r--', label = "val_loss")
plt.legend()
plt.show()
```

```python
##LSTM으로 학습해보겠습니다.
model = tf.keras.Sequential([
    tf.keras.layers.LSTM(units=30, return_sequences = True, input_shape = [100,2]),
    tf.keras.layers.LSTM(units=30),
    tf.keras.layers.Dense(1)
])
model.compile(optimizer='adam', loss='mse')
model.summary()
```

```python
history=model.fit(X[:2560], Y[:2560], epochs=100, validation_split=0.25)
```

```python
plt.plot(history.history["loss"], 'bo', label="loss")
plt.plot(history.history["val_loss"], 'r--', label = "val_loss")
plt.legend()
plt.show()
##val_loss와 loss가 급격하게 줄어들어 학습이 잘 된 것을 확인할 수 았습니다. 
```

```python
model.evaluate(X[2560:], Y[2560:])
prediction = model.predict(X[2560:2560+5])
for i in range(5):
    print(Y[2560+i], '\t', prediction[i][0], '\tdiff:', abs(prediction[i][0] - Y[2560+i]))
    
prediction = model.predict(X[2560:])
cnt = 0
for i in range(len(prediction)):
    if abs(prediction[i][0] - Y[2560+i]) > 0.04:
        cnt += 1
print('correctness:', (440 - cnt) / 440 * 100, '%')

##오차도 0에 가깝고 정확도도 매우 높은 것을 확인할 수 있습니다.
```

```python
##GRU
model = tf.keras.Sequential([
    tf.keras.layers.GRU(units=30, return_sequences = True, input_shape = [100,2]),
    tf.keras.layers.GRU(units=30),
    tf.keras.layers.Dense(1)
])
model.compile(optimizer='adam', loss='mse')
model.summary()
```

```python
history=model.fit(X[:2560], Y[:2560], epochs=100, validation_split=0.2)
```

```python
plt.plot(history.history["loss"], 'bo', label="loss")
plt.plot(history.history["val_loss"], 'r--', label = "val_loss")
plt.legend()
plt.show()
##LSTM과 비교해보면 더 빠른 epochs에서 값이 급격하게 줄어들고 값의 변화도 더 안정적인 것을 확인할 수 있습니다.
```

```python
model.evaluate(X[2560:], Y[2560:])
prediction = model.predict(X[2560:2560+5])
for i in range(5):
    print(Y[2560+i], '\t', prediction[i][0], '\tdiff:', abs(prediction[i][0] - Y[2560+i]))
    
prediction = model.predict(X[2560:])
cnt = 0
for i in range(len(prediction)):
    if abs(prediction[i][0] - Y[2560+i]) > 0.04:
        cnt += 1
print('correctness:', (440 - cnt) / 440 * 100, '%')
```
