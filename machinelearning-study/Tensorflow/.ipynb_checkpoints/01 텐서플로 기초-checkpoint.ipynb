```python
import tensorflow as tf
tf.__version__
```

```python
##신경망은 많은 숫자로 구성된 행렬입니다. 이때 행렬을 구성하는 숫자는 처음에는 랜덤하게 지정합니다.
##이처럼 신경망의 초깃값을 지정해주는 것을 초기화라고 합니다. 
##랜덤으로 지정하는 방법은 다음과 같습니다.

rand = tf.random.uniform([1], 0,1)
print(rand)
```

```python
##uniform은 균일 분포를 통해 난수를 생성한다는 것이고, [1]은 shape를 의마하는데 이는 차원의 수를 나타내는 값입니다.
##0,1은 최소와 최대 값을 나타냅니다. 

rand = tf.random.normal([4], 0,1)
print(rand)
##normal은 정규분포로 난수를 생성한다는 뜻이고, 0,1은 각각 평균과 표준편차를 의미합니다. 
```

```python
'''
신경망의 가장 기본적인 구성요소는 뉴런입니다. 뉴런은 퍼셉트론이라고도 하는데, 입력을 받아서 계산 후 출력을 반환하는 단순한
구조입니다. 신경망은 뉴런이 여러개가 모여 layer를 이루고, 이 layer들이 여러개 모인 형태입니다. 

뉴런은 입력, 가중치, 활성화함수로 구성됩니다. 
간단한 형태의 뉴런은 입력x가중치를 활성화 함수에 적용하여 결과를 출력하는 형식입니다. 
뉴런에서 학습할 때 변하는 것은 가중치 입니다. 초기화를 통해 랜덤값을 준 후, 학습을 통해 가중치를 변화시켜 일정한 값으로
수렴시키도록 합니다. 

활성화 함수로는 sigmoid 와 ReLU가 있습니다.  sigmoid는 0~1사이 값으로, ReLU는 음수는 0, 양수는 y=x를 이루도록 바꿔주는 합수입니다.
sigmoid를 활성화 함수로 사용해보도록 하겠습니다. 
'''
```

```python
import math

def sigmoid(x):
    return 1/(1+math.exp(-x))
```

```python
##입력이 0일 때 출력이 기대 출력이 1이 되는 뉴런을 만들어보겠습니다.
x = 1
y = 0
w = tf.random.normal([1], 0,1)
output = sigmoid(x*w)
print(output)
```

```python
##실제 출력과 기대 출력의 차이인 0 - 0.3490301763808875가 error입니다.
##뉴런의 학습은 error가 0에 가까워지게 해서 출력으로 기댓값에 가까운 값을 얻는 것입니다.
##w를 변화시켜야 하는데 이때 사용하는 알고리즘이 경사 하강법(Gradient Descent)을 이용합니다.
##경사하강 알고리즘은 다음과 같습니다
## w = w + x*a*error
##여기서는 a의 값을 0.1로 하겠습니다,
```

```python
for i in range(1001):
    output = sigmoid(x*w)
    error = y-output
    w = w+ x*0.1*error
    if i%100==0:
        print("{},{:.4f},{:.4f}".format(i, error, output))
```

```python
##반복적인 학습을 통해 error 값이 0에 가까워지는 것을 확인할 수 있습니다. 
##이번에는 0을 넣었을 때 1을 출력하는 뉴런을 만들어 보도록 하겠습니다.

x = 0
y = 1
w = tf.random.normal([1], 0,1)

for i in range(1001):
    output = sigmoid(x*w)
    error = y - output
    w = w+ x*0.1*error
    if i%100==0:
        print("{},{:.4f},{:.4f}".format(i, error, output))
```

```python
##error가 0.5로 고정된 것을 확인할 수 있습니다.
##이는 x가 0이기 때문에 w = w + x*a*error에서 w가 update가 되지 못합니다. 
##이렇게 가중치가 학습을 못하는 것을 방지하기 위해 편향(bias)라는 것을 뉴런에 넣어줍니다. 편향은 늘 한쪽으로 치우친 고정된 값입니다.
##편향을 추가하여 다시 학습을 해보도록 하겠습니다.

x = 0
y = 1
w = tf.random.normal([1], 0,1)
b = tf.random.normal([1], 0,1)

for i in range(1001):
    output = sigmoid(x*w + 1*b)
    error = y - output
    w = w + x*0.1*error
    b = b + 1*0.1*error
    if i%100==0:
        print("{},{:.4f},{:.4f}".format(i, error, output))
```

```python
##편향 값을 주면 학습이 잘 진행되어 error가 0으로 가까워 지는 것을 확인할 수 있습니다. 
```

```python
##AND 연산을 수행하는 뉴런을 만들어보겠습니다.
import numpy as np
x = np.array([[1,1],[1,0],[0,1],[0,0]])
y = np.array([[1],[0],[0],[0]])

w = tf.random.normal([2], 0,1)
b = tf.random.normal([1], 0,1)
b_x = 1

for i in range(2001):
    error_sum = 0
    for j in range(4):
        output = sigmoid(np.sum(x[j]*w) + b*b_x)
        error = y[j][0] - output
        w = w + x[j]*0.1*error
        b = b + b_x*0.1*error
        error_sum += error
    if i%100==0:
        print("{},{:.4f}".format(i, error_sum))
```

```python
for i in range(4):
    print("X:", x[i], "Y:", y[i], "Output:", sigmoid(np.sum(x[i]*w)+b))
```

```python
##이번에는 OR기능을 하는 뉴런을 만들어 보도록 하겠습니다.
x = np.array([[1,1],[1,0],[0,1],[0,0]])
y = np.array([[1],[1],[1],[0]])

w = tf.random.normal([2], 0,1)
b = tf.random.normal([1], 0,1)
b_x = 1

for i in range(2001):
    error_sum = 0
    for j in range(4):
        output = sigmoid(np.sum(x[j]*w) + b*b_x)
        error = y[j][0] - output
        w = w + x[j]*0.1*error
        b = b + b_x*0.1*error
        error_sum += error
    if i%100==0:
        print("{},{:.4f}".format(i, error_sum))
```

```python
for i in range(4):
    print("X:", x[i], "Y:", y[i], "Output:", sigmoid(np.sum(x[i]*w)+b))
```

```python
###이번에는 XOR을 뉴런으로 구현해보도록 하겠습니다.
x = np.array([[1,1],[1,0],[0,1],[0,0]])
y = np.array([[0],[1],[1],[0]])

w = tf.random.normal([2], 0,1)
b = tf.random.normal([1], 0,1)
b_x = 1

for i in range(2001):
    error_sum = 0
    for j in range(4):
        output = sigmoid(np.sum(x[j]*w) + b*b_x)
        error = y[j][0] - output
        w = w + x[j]*0.1*error
        b = b + b_x*0.1*error
        error_sum += error
    if i%100==0:
        print("{},{:.4f}".format(i, error_sum))
```

```python
for i in range(4):
    print("X:", x[i], "Y:", y[i], "Output:", sigmoid(np.sum(x[i]*w)+b))
```

```python
##보시는 것처럼 output은 0.5 근처에 머물러 있어 학습이 제대로 안된 것을 확인할 수 있습니다.
##해결책은 여러 개의 퍼셉트론을 사용하는 것입니다. 3개의 퍼셉트론을 사용해보겠습니다.
```

```python
x = np.array([[1,1],[1,0],[0,1],[0,0]])
y = np.array([[0],[1],[1],[0]])

model = tf.keras.Sequential([
    tf.keras.layers.Dense(units = 2, activation='sigmoid', input_shape=(2,)),
    tf.keras.layers.Dense(units = 1, activation='sigmoid')
])
model.compile(optimizer = tf.keras.optimizers.SGD(lr=0.1), loss='mse')
model.summary()
```

```python
'''
model은 딥러닝 계산을 간편하게 하기 위한 추상적인 클래스입니다. model은 딥러닝을 계산하는 가장 핵심적인 단위입니다.
Sequential은 레이어를 일직선으로 배치하겠다는 뜻입니다.
layers.Dense는 사용하는 레이러를 정의하는 명령입니다.
units는 뉴런의 수, activation은 활성화함수입니다. input_shape는 입력 데이터의 차원입니다.
summary를 보시면 Param이 6, 3인 것을 확인할 수 있는데 이는 각 레이어가 기본적으로 편향을 포함하고 있다는 뜻입니다.
compile는 모델이 실제로 동작할 수 있도록 준비하는 명령으로, optimizer는 딥러닝의 학습식을 정의하는 부분이고 여기서는
SGD, 확률적 경사 하강법을 사용하였습니다. loss로는 편균 제곱 오차를 사용하였습니다. 
학습은 fit, 예측은 predict 메서드를 사용합니다.
'''
```

```python
history = model.fit(x, y, epochs=2000, batch_size = 1)
```

```python
model.predict(x)
##앞선 결과와 다르게 얘측을 옳바르게 한 것을 확인할 수 있습니다.
```
