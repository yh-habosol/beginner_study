```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


##load_dataset은 Seabon 내장 데이터를 데이터프레임으로 가져오는 함수입니다.

titanic = sns.load_dataset('titanic')
print(titanic.head())
print("\n")
print(titanic.info())
```

```python
##regplot()함수는 서로 다른 2개의 연속변수 사이의 산점도를 그리고 선형회귀분석에 의한 회귀선을 함께 나타내는 함수

sns.set_style("darkgrid")
fig = plt.figure(figsize = (15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.regplot(x='age',
            y='fare',
            data = titanic,
            ax = ax1)


##fit_reg옵션을 False로 줌으로써 선형회귀선을 표시하지 않았습니다.

sns.regplot(x='age',
            y='fare',
            data = titanic,
            ax = ax2,
            fit_reg = False)
plt.show()
```

```python
#단변수 데이터의 분포를 확인할 때는 displot()함수를 사용합니다. 기본값으로 히스토그램과 커널 밀도 함수를 출력합니다.
#이때 커널 밀도 함수란, 그래프와 x축 사이의 면적이 1이 되도록 그리는 밀도 분포 함수입니다.

fig = plt.figure(figsize = (15,5))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

sns.distplot(titanic['fare'],ax = ax1)

sns.distplot(titanic['fare'],ax = ax2, hist = False)

sns.distplot(titanic['fare'],ax = ax3, kde = False)

plt.show()
```

```python
#히트맵을 그리기 위해 heatmap()메소드를 제공합니다. 
#2개의 범주형 데이터를 각각 x, y축에 놓고 매트릭스 형태로 분류힙니다. 


##데이터프레임을 피벗테이블로 정리하는 코드로, 이 부분은 나중에 다시 다루도록 하겠습니다.
#aggfunc='size'는 데이터 값의 크기를 기준으로 집계한다는 뜻입니다.

table = titanic.pivot_table(index=['sex'], columns = ['class'], aggfunc='size')

sns.heatmap(table,
            annot = True, fmt = 'd',                         ## 데이터 값 표시 여부, 정수형 포맷
            cmap = 'YlGnBu', 
            linewidth=.5,
            cbar = False)                                    ##color bar 표시 여부
```

```python
##stripplot()함수와 swarmplot()함수를 통해 범주형 변수에 들어 있는 각 범주별 데이터의 분포를 확인할 수 있습니다.
##swarmplot()함수는 데이터의 분산까지 고려하여 데이터 포인트가 서로 중복되지 않도록 그립니다.

sns.set_style('whitegrid')

fig = plt.figure(figsize = (15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.stripplot(x="class",
              y="age",
              data = titanic,
              ax = ax1)

sns.swarmplot(x="class",
              y="age",
              data = titanic,
              ax = ax2)
```

```python
##barplot()함수는 막대 그래프를 그리는 함수입니다. hue라는 옵션이 있는데 이 값을 추가하여 누적으로 출력 할 수 있습니다.

fig = plt.figure(figsize = (15,5))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)


#hue 옵션이 없는 경우 survived를 단순히 male, female로 나눔
sns.barplot(x='sex', y='survived', data = titanic, ax = ax1)

#hue = 'class'일 경우 male에서 survived를 class에 따라, female에서 survived를 class에 따라서 누적으로 출력함
sns.barplot(x='sex', y='survived', data = titanic, ax = ax2, hue = 'class')

#dodge옵션을 통해 stack형식으로 누적으로 출력함
sns.barplot(x='sex', y='survived', data = titanic, ax = ax3, hue = 'class', dodge = False)
```

```python
##countplot()는 각 범주에 속하는 데이터의 개수를 막대 그래프로 나타내는 함수입니다.

fig = plt.figure(figsize = (15,5))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)


sns.countplot(x='class', palette='Set1', data = titanic, ax = ax1)

sns.countplot(x='class', palette='Set2', data = titanic, ax = ax2, hue = 'who')

sns.countplot(x='class', palette='Set3', data = titanic, ax = ax3, hue = 'who', dodge = False)

plt.show()
```

```python
##박스 플롯은 범주형 데이터 분포와 주요 통계 지표를 함께 제공함. 여기서는 분산의 정도를 알 수 없기 때문에 커널 밀도 함수 그래프를 
##y 축 방향으로 추가하여 바이올린 그래프를 그리는 경우도 있음. 박스 플롯은 boxplot(), 바이올린 그래프는 violinplot()함수임.

sns.set_style('whitegrid')

fig = plt.figure(figsize = (15,10))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

sns.boxplot(x = 'alive', y = 'age', data = titanic, ax = ax1)

sns.boxplot(x = 'alive', y = 'age', data = titanic, ax = ax2, hue = 'sex')

sns.violinplot(x = 'alive', y = 'age', data = titanic, ax = ax3)

sns.violinplot(x = 'alive', y = 'age', data = titanic, ax = ax4, hue = 'sex')
```

```python
##jointplot()함수는 산점도를 기본으로 표시하고, x-y축에 각 변수에 대한 히스토그램을 동시에 보여줌.

j1 = sns.jointplot(x = 'fare', y = 'age', data = titanic)

#kind = 'reg' 옵션을 통해 회귀선 표시
j2 = sns.jointplot(x = 'fare', y = 'age', data = titanic, kind = 'reg')

#kind = 'hex'옵션을 통해 육각 그래프 그림
j3 = sns.jointplot(x = 'fare', y = 'age', data = titanic, kind = 'hex')

#kind='kde' 옵션을 통해 커널 밀집 그래프 그림
j4 = sns.jointplot(x = 'fare', y = 'age', data = titanic, kind = 'kde')
```

```python
##FacetGrid()함수는 행, 열 방향으로 서로 다른 조건을 적용하여 여러 개의 플롯을 만듬. 각 서브 플롯에 적용할 그래프 종류를
##map() 메소드를 통해 그리드 객체에 전달함.

g = sns.FacetGrid(data = titanic, col = 'who', row = 'survived')

g = g.map(plt.hist, 'age')

##열 방향으로는 who 값들이, 행 방향으로는 survived값을이 있는 것을 알 수 있고 각각의 관계에 대한 히스토그램이 그려짐.
```

```python
##pairplot()함수는 인자로 전달되는 데이터프레임의 열의 모든 짝간의 관계를 그래프로 출력함.
##같은 변수끼리는 히스토그램을, 다른 변수 간에는 산점도를 그림.

titanic_pair = titanic[['age', 'pclass','fare']]

g = sns.pairplot(titanic_pair)
```
