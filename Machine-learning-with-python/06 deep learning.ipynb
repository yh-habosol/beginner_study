```python
import mglearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
'''
신경망이라 알려진 알고리즘들은 딥러닝이란 이름으로 다시 주목받고 있습니다.
딥러닝 알고리즘의 출발점이며 비교적 간단한 모델인 다층 퍼셉트론에 대해 다뤄보겠습니다.
MLP는 여러 단계를 거쳐 결정을 만들어내는 선형 모델의 일반화된 모습입니다.
그림으로 보면 다음과 같습니다.
'''
```

```python
display(mglearn.plots.plot_logistic_regression_graph())
```

```python
#왼쪽 노드는 입력 특성을, 연결선은 학습된 계수를 표현하고 오른쪽 노드는 입력의 가중치 합, 즉 출력을 나타냅니다.
#MLP에서는 가중치 합을 만들어 내는 과정이 여러번 반복됩니다.
display(mglearn.plots.plot_single_hidden_layer_graph())
```

```python
##이런 모델을 더 강력하게 만들기 위해 각 은닉 유닛의 가중치 합을 계산한 후 결과에 비선형 함수인 ReLU나 tanh를 적용합니다.
##이 함수 결과의 가중치 합을 계산하여 출력을 만듭니다.
line = np.linspace(-3,3,100)
plt.plot(line,np.tanh(line), label="tanh")
plt.plot(line, np.maximum(line, 0), linestyle='--', label = "relu")
plt.legend(loc="best")
plt.xlabel('x')
plt.ylabel("relu(x), tahnh(x)")
```

```python
##위 신경망 그림에서 tanh를 이용하면 입력과 은닉 유닛 사이에 가중치 합을 tanh에 놓고, 그 결과와 은닉과 출력 사이의 가중치 합을
##구하면 출력이 나옵니다.
##이때 은닉층의 유닛 개수가 중요한데 개수를 늘리거나 새로운 은닉층을 만들 수 있습니다.

mglearn.plots.plot_two_hidden_layer_graph()

#이처럼 많은 은닉층이 생기면서 딥러닝이 생기게 되었습니다.
```

```python
##신경망을 튜닝해보도록 하겠습니다.
##MLP는 기본값으로 은닉 유닛을 100개 가지고 solver로는 lbfgs를 사용하였습니다. 기본 비선형 함수는 ReLU를 사용합니다. 
##결정 경계를 매끄럽게 하기 위해서는 은닉 유닛을 추가하거나, tanh를 사용하거나, 은닉층을 추가하는 방법이 있습니다.
##각각의 경우에 대해 결정 경게를 보여드리도록 하겠습니다.

from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split

X,y  = make_moons(n_samples = 100, noise = 0.25, random_state = 3)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, random_state = 42)


###기본 설정
mlp = MLPClassifier(solver='lbfgs', random_state = 0)
mlp.fit(X_train, y_train)
mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha = .3)
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train)
```

```python
##은닉 유닛 개수 10개
mlp = MLPClassifier(solver='lbfgs', random_state = 0, hidden_layer_sizes = [10])
mlp.fit(X_train, y_train)
mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha = .3)
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train)
```

```python
##10개 은닉 유닛 층 2개
mlp = MLPClassifier(solver='lbfgs', random_state = 0, hidden_layer_sizes = [10, 10])
mlp.fit(X_train, y_train)
mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha = .3)
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train)
```

```python
##tanh를 사용, 10개 유닛 은닉층 2개
mlp = MLPClassifier(solver='lbfgs', random_state = 0, hidden_layer_sizes = [10,10], activation = 'tanh')
mlp.fit(X_train, y_train)
mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha = .3)
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train)
```

```python
##L2규제를 이용하여 가중치를 0에 가깝게 하여 모델의 복잡도를 제어할 수 있습니다
##여기서도 복잡도를 제어하는 매개변수는 alpha값입니다. 

fig, axes = plt.subplots(2, 4, figsize=(20, 8))
for axx, n_hidden_nodes in zip(axes, [10, 100]):
    for ax, alpha in zip(axx, [0.0001, 0.01, 0.1, 1]):
        mlp = MLPClassifier(solver='lbfgs', random_state=0,
                            hidden_layer_sizes=[n_hidden_nodes, n_hidden_nodes],
                            alpha=alpha, max_iter=1000)
        mlp.fit(X_train, y_train)
        mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=.3, ax=ax)
        mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train, ax=ax)
        ax.set_title("n_hidden=[{}, {}]\nalpha={:.4f}".format(
                      n_hidden_nodes, n_hidden_nodes, alpha))
```

```python
##신경망에서는 학습을 시작하기 전에 가중치를 무작위로 설정하며 무작위한 초기화가 학습에 영향을 많이 줍니다. 
##매개변수가 같아도 초깃값이 다른 경우를 그래프로 살펴보도록 하겠습니다.

fig, axes = plt.subplots(2, 4, figsize=(20, 8))
for i, ax in enumerate(axes.ravel()):
    mlp = MLPClassifier(solver='lbfgs', random_state=i,
                        hidden_layer_sizes=[100, 100])
    mlp.fit(X_train, y_train)
    mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=.3, ax=ax)
    mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train, ax=ax)
```

```python
##실제 데이터셋에서 성능을 확인해 보도록 하겠습니다.
import warnings
warnings.filterwarnings("ignore")

from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()

X_train,X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state = 0)

###정확도를 높이기위해 각 데이터들을 표준화 하도록 하겠습니다.

mean_on_train = X_train.mean(axis = 0)
std_on_train = X_train.std(axis = 0)

X_train_scaled = (X_train-mean_on_train)/std_on_train
X_test_scaled = (X_test-mean_on_train)/std_on_train

mlp = MLPClassifier(max_iter = 1000, random_state = 0)
mlp.fit(X_train_scaled, y_train)

print(mlp.score(X_train_scaled, y_train))
print(mlp.score(X_test_scaled, y_test))
```

```python
##훈령과 테스트 사이의 성능 차이가 있으므로 규제를 강하게 해보겠습니다.

mlp = MLPClassifier(max_iter = 1000, random_state = 0, alpha = 1)
mlp.fit(X_train_scaled, y_train)

print(mlp.score(X_train_scaled, y_train))
print(mlp.score(X_test_scaled, y_test))
```
