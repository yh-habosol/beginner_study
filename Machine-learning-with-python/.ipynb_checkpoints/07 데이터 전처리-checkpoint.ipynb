```python
##MinMaxScaler     .
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state = 1)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
print("  : \n", X_train.min(axis= 0))
print("  : \n",X_train.max(axis= 0))
print("  : \n", X_train_scaled.min(axis = 0))
print("  : \n",X_train_scaled.max(axis = 0))
```

```python
##MinMaxScaler   0~1      .
##  .

X_test_scaled = scaler.transform(X_test)
print("  : \n", X_test_scaled.min(axis = 0))
print("  : \n",X_test_scaled.max(axis = 0))
```

```python
##     .
from sklearn.svm import SVC

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state = 0)

svm = SVC(C = 100)
svm.fit(X_train, y_train)
print("{:.4f}".format(svm.score(X_test, y_test)))
print('\n')

scaler = MinMaxScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
svm.fit(X_train_scaled, y_train)
print("{:.4f}".format(svm.score(X_test_scaled, y_test)))

##     .
```
