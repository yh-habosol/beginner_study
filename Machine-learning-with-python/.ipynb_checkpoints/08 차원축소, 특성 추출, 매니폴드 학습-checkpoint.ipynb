```python
import mglearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
##주성분 분석(PCA)은 특성들이 통계적으로 상관관계가 없도록 데이터셋을 회전시키는 기술입니다.
##다음 그래프는 인위적으로 만든 2차원 데이터셋을 사용하여 PCA효과를 나타낸 것입니다.
import warnings
warnings.filterwarnings("ignore")
mglearn.plots.plot_pca_illustration()
```

```python
'''
PCA알고리즈은 다음과 같습니다.
첫번째 그래프에서 성분 1이라고 쓰여져있는 분산이 가장 큰 방향을 찾습니다. 그 후, 이 방향과 수직인 방향들 중 가장 많은 정보를 담은
방향을  찾습니다. 2차원에서는 직각인 방향이 하나이지만, 고차원에서는 무수히 많습니다.
이렇게 찾은 방향을 데이터에 있는 주된 분산의 방향이라고 해서 주성분이라고 합니다. 보통 주성분은 원본 특성의 개수만큼 있습니다.

두번째 그래프는 두 성분을 각각 x,y축으로 하도록 회전시킨 결과입니다. 이 두축은 연관되어 있지 않으므로 상관관계 행렬이 대각선 방향을
제외하고는 0아 됩니다.
PCA는 주성분의 일부만 남기는 차원 축소 용도로 사용할 수 있습니다.
세번쨰 그래프처럼 첫번째 주성분만 남길 수 있습니다. 그 후 다시 회전하면 마지막 그래프처럼 됩니다. 즉, PCA알고리즘을 통해
특성 하나에 대한 그래프가 출력된 것을 확인할 수 있습니다.
'''
```

```python
'''
PCA는 고차원 데이터의 시각화에 많이 쓰입니다.
유방암 데이터셋은 산점도 행렬로 보기 힘듭니다. 따라서 이경우에는 아래 그래프와 같이 히스토그램으로 보는 것이 유리합니다.
'''
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()

fig, axes = plt.subplots(15, 2, figsize=(10, 20))
malignant = cancer.data[cancer.target == 0]
benign = cancer.data[cancer.target == 1]

ax = axes.ravel()

for i in range(30):
    _, bins = np.histogram(cancer.data[:, i], bins=50)
    ax[i].hist(malignant[:, i], bins=bins, color=mglearn.cm3(0), alpha=.5)
    ax[i].hist(benign[:, i], bins=bins, color=mglearn.cm3(2), alpha=.5)
    ax[i].set_title(cancer.feature_names[i])
    ax[i].set_yticks(())
ax[0].set_xlabel("특성 크기")
ax[0].set_ylabel("빈도")
ax[0].legend(["악성", "양성"], loc="best")
fig.tight_layout()
```

```python
'''
히스토그램을 통해 특성들이 클래스별로 어떻게 분포되어 있는지를 알려주고, 이를 통해 어떤 특성이 양성과 악성 샘플을 구분하는 데
더 좋은지 가늠해볼 수 있습니다.
하지만 이 그래프는 특성 간의 상호작용이나 상호작용이 클래스와 어떤 관련이 있는지 알려주지 않습니다. 
PCA를 사용하면 주요 상호작용을 찾아낼 수 있어 더 나은 그림을 만들 수 있습니다.
'''
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

scaler = StandardScaler()
scaler.fit(cancer.data)
X_scaled = scaler.transform(cancer.data)

pca = PCA(n_components = 2)
pca.fit(X_scaled)

X_pca = pca.transform(X_scaled)

plt.figure(figsize=(8, 8))
mglearn.discrete_scatter(X_pca[:, 0], X_pca[:, 1], cancer.target)
plt.legend(["악성", "양성"], loc="best")
plt.gca().set_aspect("equal")
plt.xlabel("첫 번째 주성분")
plt.ylabel("두 번째 주성분")
```

```python
##PCA의 단점은 그래프의 두 축을 설명하기가 쉽지 않습니다. 
##주성분은 다음과 같습니다.
pca.components_
```

```python
##행은 중요도 순으로 정렬되고 각 열은 데이터의 열과 대응됩니다.
plt.matshow(pca.components_, cmap='viridis')
plt.yticks([0, 1], ["첫 번째 주성분", "두 번째 주성분"])
plt.colorbar()
plt.xticks(range(len(cancer.feature_names)),
           cancer.feature_names, rotation=60, ha='left')
plt.xlabel("특성")
plt.ylabel("주성분")

##히트맵으로 나타냈을 때 모든 특성이 섞여 있기 때문에 축이 가지는 의미를 설명하기 쉽지 않음을 알 수 있습니다.
```

```python
'''
다음은 비음수 행렬 분해(NMF)에 대해 다뤄보겠습니다.
NMF는 유용한 특성을 뽑아내기 위한 또 다른 비지도 학습 알고리즘입니다. 
PCA와 비슷하고 차원 축소에도 사용합니다.
차이점은, NMF에서는 음수가 아닌 성분과 계수 값을 찾습니다. 즉, 주성분과 계수가 모두 0보다 같거나 커야합니다. 
'''
mglearn.plots.plot_nmf_illustration()
```

```python
'''
왼쪽은 성분이 2개인 NMF로, 알고리즘은 데이터의 각 특성의 끝에 위치한 포인트를 가리키는 방향을 선택합니다. 
하나의 성분만 사용한다면, 데이터를 가장 잘 표현할 수 있는 평균으로 향하는 성분을 만듭니다. 
PCA와는 반대로 성분 개수를 줄이면 특정 방향이 제거되는 것뿐만 아니라 전체 성분이 아얘 바뀝니다.

PCA는 재구성 측면에서 최선의 방향을 찾기 때문에 NMF에 비해 성능이 좋습니다.
따라서 NMF는 데이터를 인코딩 하거나 재구성하는 용도로 사용하기보다는 주로 데이터에 있는 유용한 패턴을 찾는 데 사용합니다.
패턴을 추출하는 것은 소리, 유전자 표현, 텍스트 데이터처럼 덧붙이는 구조를 가진 데이터에 적합합니다.
'''
S = mglearn.datasets.make_signals()
plt.figure(figsize=(6,1))
plt.plot(S, '-')
plt.xlabel("time")
plt.ylabel("signal")
```

```python
##원본신호를 볼 수 없으므로 합쳐진 신호를 분해해서 원본 신호를 복원해야 합니다. 

from sklearn.decomposition import NMF

##원본 데이터를 사용해 100개의 측정 데이터를 만듭니다.
A = np.random.RandomState(0).uniform(size=(100, 3))
X = np.dot(S, A.T)
print("측정 데이터 형태:", X.shape)
```

```python
##NMF를 사용해 세 개의 신호를 복원합니다.

nmf = NMF(n_components=3, random_state=42, max_iter=1000, tol=1e-2)
S_ = nmf.fit_transform(X)
print("복원한 신호 데이터 형태:", S_.shape)
```

```python
pca = PCA(n_components=3)
H = pca.fit_transform(X)
```

```python
models = [X, S, S_, H]
names = ['측정 신호 (처음 3개)',
         '원본 신호',
         'NMF로 복원한 신호', 
         'PCA로 복원한 신호']

fig, axes = plt.subplots(4, figsize=(8, 4), gridspec_kw={'hspace': .5},
                         subplot_kw={'xticks': (), 'yticks': ()})

for model, name, ax in zip(models, names, axes):
    ax.set_title(name)
    ax.plot(model[:, :3], '-')
    ax.margins(0)
    
###NMF는 원본 신호를 잘 복원했지만 PCA는 실패했음을 알 수 있습니다. 
```

```python
##다음은 매니폴드학습에 대해 다뤄보도록 하겠습니다.
'''
매이폴드 학습은 복잡한 매핑을 만들어 시각화를 제공합니다. 특히 t-SNE 알고리즘을 많이 사용합니다.

매니폴드 학습의 목적은 시각화입니다. 또한 새로운 데이터에는 적용할 수 없고 훈련했던 데이터만 변환할 수 있습니다.
그래서 탐색적 데이터 분석에 유용하지만 지도 학습으로는 사용하지 않습니다.
'''
```

```python
from sklearn.datasets import load_digits
digits = load_digits()

fig, axes = plt.subplots(2, 5, figsize=(10, 5),
                         subplot_kw={'xticks':(), 'yticks': ()})
for ax, img in zip(axes.ravel(), digits.images):
    ax.imshow(img)
```

```python
##PCA를 이용하여 2차원으로 축소해 시각화해보겠습니다.
pca = PCA(n_components=2)
pca.fit(digits.data)
# 처음 두 개의 주성분으로 숫자 데이터를 변환합니다
digits_pca = pca.transform(digits.data)
colors = ["#476A2A", "#7851B8", "#BD3430", "#4A2D4E", "#875525",
          "#A83683", "#4E655E", "#853541", "#3A3120","#535D8E"]
plt.figure(figsize=(10, 10))
plt.xlim(digits_pca[:, 0].min(), digits_pca[:, 0].max())
plt.ylim(digits_pca[:, 1].min(), digits_pca[:, 1].max())
for i in range(len(digits.data)):
    # 숫자 텍스트를 이용해 산점도를 그립니다
    plt.text(digits_pca[i, 0], digits_pca[i, 1], str(digits.target[i]),
             color = colors[digits.target[i]],
             fontdict={'weight': 'bold', 'size': 9})
plt.xlabel("첫 번째 주성분")
plt.ylabel("두 번째 주성분")
```

```python
##같은 데이터 셋에 t-SNE를 적용해보겠습니다.
from sklearn.manifold import TSNE
tsne = TSNE(random_state=42)
# TSNE에는 transform 메소드가 없으므로 대신 fit_transform을 사용합니다
digits_tsne = tsne.fit_transform(digits.data)
```

```python
plt.figure(figsize=(10, 10))
plt.xlim(digits_tsne[:, 0].min(), digits_tsne[:, 0].max() + 1)
plt.ylim(digits_tsne[:, 1].min(), digits_tsne[:, 1].max() + 1)
for i in range(len(digits.data)):
    # 숫자 텍스트를 이용해 산점도를 그립니다
    plt.text(digits_tsne[i, 0], digits_tsne[i, 1], str(digits.target[i]),
             color = colors[digits.target[i]],
             fontdict={'weight': 'bold', 'size': 9})
plt.xlabel("t-SNE 특성 0")
plt.ylabel("t-SNE 특성 1")
```
