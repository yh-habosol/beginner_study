```python
##함수 매핑은 시리즈 또는 데이ㅓ프레임의 개별 원소를 특정 함수에 일대일로 대응시키는 과정을 뜻합니다.
##시리즈 객체에 apply 함수를 적용하면 인자로 전달되는 매핑함수에 시리즈의 모든 원소를 하나씩 입력하여 리턴값을 돌려받습니다.

import seaborn as sns
import pandas as pd

titanic = sns.load_dataset("titanic")
df = titanic.loc[:, ['age', 'fare']]
df['ten'] = 10
df.head()
```

```python
sr3 = df['age'].apply(lambda x: x+10)
sr3.head()
```

```python
##매핑 함수가 시리즈를 반환하는 함수라면, apply를 통해 데이터프레임이, 하나의 값을 반환하는 함수라면 시리즈 객체를 반환합니다.
titanic = sns.load_dataset("titanic")
df = titanic.loc[:, ['age', 'fare']]

def missing_value(series):
    return series.isnull()

result = df.apply(missing_value, axis = 0)
print(result.head())
print(type(result))
print("\n")

def min_max(x):
    return x.max() - x.min()

result = df.apply(min_max, axis = 0)
print(result.head())
print(type(result))
```

```python
##apply에 axis = 1로 하면, 각 행을 매핑 함수로 전달하고, 행 인덱스가 시리즈의 인덱스가 됩니다. 
titanic = sns.load_dataset("titanic")
df = titanic.loc[:, ['age', 'fare']]
df['ten'] = 10

df['add'] = df.apply(lambda x: x['age'] + x['ten'], axis = 1)
df.head()
```

```python
df = pd.read_excel('./주가데이터.xlsx')
df.head()
```

```python
##위 연월일 열을 각각 분리해보도록 하겠습니다.
##우선 자료형을 문자열로 바꾼 후, split함수를 통해 각각을 분리해야합니다.

df['연월일'] = df['연월일'].astype('str')
dates = df['연월일'].str.split("-")
print(dates)
```

```python
##get메소드를 통해 리스트의 각 원소에 접근할 수 있습니다.
df['year'] = dates.str.get(0)
df['month'] = dates.str.get(1)
df['day'] = dates.str.get(2)
df.head()
```

```python
##불린인덱스에 대해 알아보겠습니다.
##시리즈 객체에 어떤 조건식을 적용하여 참 값으로만 된 시리즈를 반환합니다. 
##어떠한 조건을 가진 데이터를 추출할 때 많이 사용됩니다.

titanic = sns.load_dataset("titanic")

mask1 = (titanic.age>=10) & (titanic.age<20)

df_teenage = titanic.loc[mask1, :]
df_teenage.head()
```

```python
##isin메소드를 적용하면 특정 값을 가진 행들을 따로 추출할 수 있습니다.
titanic = sns.load_dataset("titanic")

isin_filter = titanic['sibsp'].isin([3,4,5])
isin_df = titanic[isin_filter]
isin_df.head()
```

```python
##복잡한 데이터를 어떤 기준에 따라 여러 그룹으로 나눠서 관찰하는 것도 좋은 방법입니다.
##특정 기준으로 그룹을 분할하여 처리하는 것을 그룹연산이라고 합니다.
##데이터를 특정 조건으로 분할하는 함수를 groupby()라고 합니다. 

titanic = sns.load_dataset("titanic")
df = titanic.loc[:, ['age','sex','class','fare','survived']]

grouped = df.groupby(['class'])
print(grouped)
```

```python
for key, group in grouped:
    print(key, group)

##class는 First, Second, Third로 나뉘는데
##각각이 key값으로, 해당 key에 해당하는 데이터프레임이 group로 들어가는 것을 확인할 수 있습니다.
```

```python
for key, group in grouped:
    print(key)
    print(group.head())
    print("\n")
```

```python
##각 열에 대해 연산 메소드를 적용할 수 있는데, 연산 가능한 열에 대해서만 계산을 진행합니다. 

average = grouped.mean()
average
```

```python
##get_group()라는 메소드를 통해 특정 그룹만 선택할 수 있습니다.

group3 = grouped.get_group('Third')
group3.head()
```

```python
##groupby 함수에 멀티인덱스가 들어갈 수있습니다.
#class는 3게, sex는 2개의 데이터 값을 가지고 있으므로 이 둘을 멀티인덱스로 하면 총 6개의 그룹이 생성됩니다.

grouped_two = df.groupby(['class','sex'])
for key, group in grouped_two:
    print(key)
    print(group.head())
    print("\n")
```

```python
##멀티인덱스로된 그룹도 역시 get_group함수를 통해 특정 그룹만을 지정할 수 있습니다.
group4 = grouped_two.get_group(("Third", "female"))
group4.head()
```

```python
##분할한 그룹 객체에 대해 다양한 연산을 적용할 수 있는데 이 과정을 데이터 집계라고 합니다.

titanic = sns.load_dataset("titanic")

df = titanic.loc[:, ['age','sex','class','fare','survived']]

grouped = df.groupby('class')
std_all = grouped.std()
print(std_all)
print("\n")
std_all = grouped.fare.std()
print(std_all)
```

```python
##기본적으로 정의되는 함수가 아니라 사용자 정의 함수를 사용하거나, 여러 함수를 한번에 적용하려면 agg메소드를 사용하면 됩니다.

def min_max(x):
    return x.max() - x.min()

agg_minmax = grouped.agg(min_max)
print(agg_minmax.head())
```

```python
a = grouped.agg(['min', 'max'])

b = grouped.agg({'fare':['min', 'max'], 'age':'mean'})

print(a.head())
print("\n")
print(b.head())
```

```python
##그룹 객체에 filter메소드를 적용하면 조건이 참인 그룹만을 반환합니다.

group_filter = grouped.filter(lambda x: len(x)>200)
group_filter.head()
```

```python
##그룹객체에도 apply함수룰 적용할 수 있습니다.
a = grouped.apply(lambda x: x.describe())
a
```
